## Handle Missing Values

```python
# Count missing values in each column

missing_values = df.isnull().sum()

print(missing_values)
```

```python
# Calculate percentage of missing values

missing_percentage = (
    df.isnull().sum() / len(df)
) * 100

print(missing_percentage.sort_values(ascending=False))
```

## Fill Missing Values

```python
# Fill numerical columns with median

num_cols = df.select_dtypes(include=np.number).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Fill categorical columns with 'Unknown'

cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    df[col] = df[col].fillna("Unknown")
```

```python
# Check if missing values remain

print(df.isnull().sum())
```

## Perform Basic Operations

```python
# Select useful columns

selected_data = df[
    ['title', 'rating', 'category']
]

selected_data.head()
```

```python
# Filter products with rating greater than 4

high_rated_products = df[
    df['rating'] > 4
]

high_rated_products.head()
```

## Remove Duplicates

```python
# Count duplicate rows

print(
    "Duplicate Rows:",
    df.duplicated().sum()
)
```

## Data Cleaning for Price Columns

```python
# Remove currency symbol and commas

df['final_price'] = (
    df['final_price']
    .astype(str)
    .str.replace('‚āĻ', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.replace('"', '', regex=False)
)

# Convert to numeric datatype

df['final_price'] = pd.to_numeric(
    df['final_price'],
    errors='coerce'
)
```

## Create Derived Column

```python
# Create quantity column

df['quantity'] = 1
```

### Create total_amount

```python
# Derived column

df['total_amount'] = (
    df['final_price']
    * df['quantity']
)

df[['final_price', 'quantity', 'total_amount']].head()
```
